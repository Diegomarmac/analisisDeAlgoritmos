## Prueba
es importante que el arreglo usado como _base_est√© ordenado, el algoritmo __no__ contempla el revisar si est√° o no ordenado.  

Al ser un algoritmo recursivo algo importante es probar que el algoritmo termina, para esto tenemos 2 casos:
- El usuario indica que ya se ha encontrado el n√∫mero deseado, entonces el programa termina
- El usuario _hace trampa_, en este caso el programa repetir√° el n√∫mero a donde se le ha _acorralado_, ya que el profesor indico que no hubiera un momento de limite de intentos para terminar el programa, en este caso, el usuario deber√° indicar que se ha encontrado el n√∫mero y el programa terminar√°. En este punto, al quedar _acorralado_ en un solo n√∫mero, es el fin del del ciclo, es decir se mantendr√° ah√≠ mientras el usuario as√≠ lo desee, el loop no se mover√°. por lo que queda notorio que los limites no incrementan infinitamente.

¬ø C√∫antas Veces se har√° el loop ?
- Mejor caso: 1, en este caso el arreglo es de tama√±o 0 o 1, de modo que no puede realizar la operaci√≥n $LI + (\frac{LS-LI}{2})$ o donde al primer intento se encuentra el n√∫mero deseado.
- Caso Promedio: Aqu√≠ no importa de que lado queda el n√∫mero a buscar, lo que nos interesa es que exista una iteraci√≥n siguiente donde los subarreglos sean menores que los anteriores

Dentro de la longitud del ArregloOrdenado nuestro algoritmo se ejecutar√° de forma recursiva, primero se toma la mitad de la longitud del arreglo, llamemos a este n√∫mero n, y se pregunta si es el n√∫mero a buscar, de forma que si es, hemos encontrado el mejor caso, de otra manera tenemos 2 posibles casos que el n√∫mero a buscar sea menor o sea mayor de la mitad del arreglo.  
Si el n√∫mero a buscar es menor que el dado, entonces como ArregloOrdenado est√° ordenado, podemos seguir _partiendo a mitades_ , para este caso, actualizamos el l√≠mite superior de busqueda como el primer n√∫mero dado, es decir, el l√≠mite superior es ahora n, y volvemos a ejecutar el algoritmo pero teniendo ahora como limite superior n. En este punto es notorio que se regresa al caso base, con los valores actualziados. 
Si el n√∫mero a buscar es mayo, entonces tomamos n como el nuevo limite inferior, y ahora va desde n hasta el extremo superior, este ser√° nuestro nuevo arreglo para buscar, en este punto volvemos a realziar la busqueda de forma recursiva, con lo que regresaremos al caso base.

## El codigo

```julia
function guessTheNumber(limiteSuperior, limiteInferior)
    variableABuscar = limiteInferior + ((limiteSuperior - limiteInferior)/2) #El principio de BInary Search se basa en irnos a mitades hacia arriba o hacia abajo, usamos ese principio
    variableABuscar = trunc(Int,variableABuscar) # Con esto convertimos el resultado a Entero
    println("El n√∫mero a buscar es: ", variableABuscar, "?")
    println("Presione 'g' si el n√∫mero a buscar es m√°s grande")
    println("Presione 'b' si el n√∫mero a buscar es m√°s bajo")
    println("Presione 'c' si el n√∫mero a buscar es el correcto")
    respuesta = chomp(readline())
    while respuesta ‚àâ ["c", "b", "g"] #Que el usuario SOLAMENTE pueda ingresar las opciones que queremos
        print("Por favor seleccione solamente \"c\", \"b\", o \"g\". ")
        respuesta = chomp(readline())
    end
    if respuesta == "c" # Esto es solamente gesti√≥n de entrada, si ya se encontr√≥ el n√∫mero, pos GAME OVER
        println("Genial ya encontramos el n√∫mero!")
    elseif respuesta == "b"
        limiteSuperior = variableABuscar
        guessTheNumber(limiteSuperior, limiteInferior) # Si el n√∫meroa buscar es menor, entonces tomamos el n√∫mero dado como el limite superior para la siguiente busqueda
    else
        limiteInferior = variableABuscar
        guessTheNumber(limiteSuperior,limiteInferior) # Si el n√∫meroa buscar es mayor, entonces tomamos el n√∫mero dado como el limite inferior para la siguiente busqueda
    end
end
```
